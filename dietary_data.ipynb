```python
# Rename Day 2 columns to match Day 1 format
nutrient_day2 = nutrient_day2.rename(columns=lambda x: x.replace("DR2", "DR1") if x.startswith("DR2") else x)

# Verify new column names
print(nutrient_day2.columns)
```

```python
# Merge nutrient intake data using participant ID (SEQN)
nutrient_avg = pd.merge(nutrient_day1, nutrient_day2, on="SEQN", how="outer", suffixes=("_D1", "_D2"))
```

```python
# Compute average daily intake for participants with both days available
for col in nutrient_day1.columns:
    if col != "SEQN":  # Skip participant ID
        col_d1 = f"{col}_D1"
        col_d2 = f"{col}_D2"

        # Check if both Day 1 & Day 2 columns exist before computing mean
        if col_d1 in nutrient_avg.columns and col_d2 in nutrient_avg.columns:
            nutrient_avg[col] = nutrient_avg[[col_d1, col_d2]].mean(axis=1, skipna=True)
        elif col_d1 in nutrient_avg.columns:
            nutrient_avg[col] = nutrient_avg[col_d1]
        elif col_d2 in nutrient_avg.columns:
            nutrient_avg[col] = nutrient_avg[col_d2]
```

```python
nutrient_avg.head()
```

```python
combined_missing_values_n = nutrient_avg.isnull().mean() * 100
combined_missing_values_n 
```

```python
threshold = 50  # Percentage
columns_to_drop = combined_missing_values_n[combined_missing_values_n > threshold].index
nutrient_avg = nutrient_avg.drop(columns=columns_to_drop)
```

```python
# Select columns where missing percentage is between .87% and 13%
columns_to_filter = combined_missing_values_n[(combined_missing_values_n >= 23) & (combined_missing_values_n <= 35)].index

# Drop rows where any of the selected columns have missing values
filtered_data_n = nutrient_avg.dropna(subset=columns_to_filter)

# Check new shape after dropping rows
print(f"Original shape: {nutrient_avg.shape}")
print(f"New shape after dropping rows: {filtered_data_n.shape}")
```
